```python
import albumentations as A
import albumentations.pytorch
from PIL import Image
import numpy as np

class CLAHEAugmentation:
    ## Use albumenation
    def __init__(self, **args):
        self.transform = A.Compose([
            A.CenterCrop(320, 256),
            A.CLAHE(),
            # A.pytorch.transforms.ToTensor()
        ])

    def __call__(self, input):
        return self.transform(image = np.array(input))["image"]

transform = CLAHEAugmentation()
image_dir = ('/opt/ml/input/data/train/images/000001_female_Asian_45/incorrect_mask.jpg')
image = Image.open(image_dir)
display(image)
image_transform = transform(image)
pil_image=Image.fromarray(image_transform)
display(pil_image)
```
